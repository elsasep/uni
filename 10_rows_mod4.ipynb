```python
import pandas as pd
```

```python
# Data Lontoon Airbnb-listauksista luetaan suoraan sivustolta, ei tarvitse ladata koneelle
url = "https://data.insideairbnb.com/united-kingdom/england/london/2025-09-14/visualisations/listings.csv"
data = pd.read_csv(url)

data.head()
```

```python
print(data.columns)
```

```python
# Kerätään uuteen dataframeen tiedot hosteista, kohteiden ja arvioiden määrästä sekä keskihinnasta
host_data = data.groupby("host_id").agg(
    {
        "host_name": "first",
        "id": "count",
        "price": "mean",
        "number_of_reviews": "sum",
    }
).reset_index()

# Nimetään sarakkeet uudelleen
host_data.columns = ["host_id", "host_name", "listing_count", "average_price", "total_reviews"]
host_data.head()
```

```python
host_data.isna().sum()
```

```python
# Poistetaan rivit, joissa ei ole hintaa
host_data = host_data.dropna(subset=["average_price"])
```

```python
host_data.shape
```

```python
avg_listings = host_data["listing_count"].mean()
print(f"Keskimääräinen kohteiden määrä per host: {avg_listings:.2f}")
```

```python
avg_price = host_data["average_price"].mean()
print(f"Keskimääräinen hinta per yö: {avg_price:.2f} puntaa")
```

```python
avg_reviews = host_data["total_reviews"].mean()
print(f"Keskimääräinen arvioiden määrä per host: {avg_reviews:.0f}")
```
